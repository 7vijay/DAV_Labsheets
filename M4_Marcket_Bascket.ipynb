#What is Apriori algorithm
The apriori algorithm is the most common technique for performing market basket analysis. it is used for association rule mining, which is a rule based process used to identify co relations between items purchased by users.

#Whta sre the components of apriori algorithm
The apriori algorithm has three main components
1. Support
2. lift
3. confidence

**Support**
Support(item)=Transactions comprising the item/Total transactions


A high support value indicates that the item is present in most purchases,therefore markets should focus on it more.

**Confidence**
It tells us about the likelohood of different purchases combinations. we calculate that using the following formula:
Confidence(Bread->Milk)= Transactions comprising bread and milk / Transactions comprising bread
**Lift**
Finally Lift refers to the increase in the ratio of the sale of milk when you sell bread:
Lift= Confidence(Bread -> Milk)/Support(Bread)=0.75/1=1.3.
This Means that customers are 1.3 times more likely to buy milk if u also sell bread.

**Step 2: Reading the Dataset**

```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Groceries_dataset.csv')
```

```python
df['single_transcation']=df['Member_number'].astype(str)+'_'+df['Date'].astype(str)
df.head()
```

```python
df2 = pd.crosstab(df['single_transcation'], df['itemDescription'])
df2.head()
```

```python
def encode(item_freq):
  res=0
  if item_freq > 0:
    return res

basket_input=df2.applymap(encode)
```

**step 4: build the apriori algorithm for market basket analysis**

now, let's import the apriori algorithm from the MLXtend python package and use it to discover frequently-bought together item combinations

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```
