```python
df=pd.read_csv('/content/netflix_titles.csv',usecols=['cast'])
df.head()
```

```python
ndf=df.dropna()
ndf.head()
```

```python
text=" ".join(item for item in ndf['cast'])
print(text)
```

```python
stopwords=set(STOPWORDS)
```

```python
wordcloud=WordCloud(background_color="white").generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.margins(x=,y=0)
plt.show()
```

```python
wordcloud=WordCloud(background_color="white",max_words=100,max_font_size=100,width=300,height=200,colormap="rainbow").generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis=("off")
plt.margins(x=0,y=0)
plt.savefig("cloud.jpg",format="jpg")
```
