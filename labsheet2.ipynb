## Creating dataframes from Dictonary and series

```python
data=[{'a':1,'b':2},{'a':5,'b':10,'c':20}]
df=pd.DataFrame(data,index=['First','Second'])
print(df)
```

```python
data=[{'a':1,'b':2},{'a':5,'b':10,'c':20}]
#with 2 column indices, values same as dictonary keya
df1=pd.DataFrame(data,index=['First','Second'],columns=['a','b'])
#with 2 column indices with one index with other name
df2=pd.DataFrame(data,index=['First','Second'],columns=['a','b1'])
print(df1)
print(df2)
```

# Column Addition

```python
d={'one': pd.Series([1,2,3],index=['a','b','c']),
   'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df['one'])
#Adding a new column to an existing Dataframe object with column label by passing new series
print("Adding a new column by passing a Series : ")
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)
print("Adding a new column by using existing column in dataframe : ")
df['four']=df['one']+df['three']
print(df)
```

```python
d={'one': pd.Series([1,2,3],index=['a','b','c']),
   'two': pd.Series([1,2,3,4],index=['a','b','c','d']),
   'three': pd.Series([10,20,30],index=['a','b','c'])}
df=pd.DataFrame(d)
print("Our Dataframe is : ")
print(df)
#using del for delition
print("Deleating the first column using del function")
del(df['one'])
print(df)
#using pop function
print("Deleating the another column using pop function")
df.pop('two')
print(df)
```

```python
d={'one': pd.Series([1,2,3],index=['a','b','c']),
   'two': pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df.loc['b'])
```

```python
d={'one': pd.Series([1,2,3],index=['a','b','c']),
   'two': pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df.iloc[2])
```

```python
d={'one': pd.Series([1,2,3],index=['a','b','c']),
   'two': pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df.iloc[2:4])
```

## Addition of Rows

```python
df=pd.DataFrame([[1,2],[3,4]],columns=['a','b'])
df2=pd.DataFrame([[5,6],[7,8]],columns=['a','b'])
df=df.append(df2)
print(df)
```

## Deletion of rows

```python
df=pd.DataFrame([[1,2],[3,4]],columns=['a','b'])
df2=pd.DataFrame([[5,6],[7,8]],columns=['a','b'])
df=df.append(df2)
#drop rows with label 0
df=df.drop(0)
print(df)
```
