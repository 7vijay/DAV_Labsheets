```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import seaborn as sns
```

```python
df1 = sns.load_dataset('titanic')
print(df1.head)
```

```python
sns.barplot(x = 'survived', y='class', data = df1,palette ='plasma',hue='class').set_title('bar Plot')
```

```python
sns.barplot(x = 'class', y='fare', data = df1,palette ='plasma',hue='class', estimator = np.std).set_title('bar Plot')
```

```python
sns.violinplot(x = 'survived', y ='embark_town', data= df1,hue='embark_town', split= True).set_title('Violin Plot')
```

```python
g=sns.pairplot(df1,hue='survived',palette='coolwarm')
g.fig.suptitle("Pair Plot 1")
g.add_legend()
```
